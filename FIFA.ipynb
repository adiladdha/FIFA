```python
player_row = fifa1_percentiles[fifa1_percentiles['Known As'] == best_player]
attribute_values1 = player_row.iloc[:, 1:].values.flatten()

# Print the attribute values

attribute_values1     #percentiles from same position
```

```python
player_row2 = fifa_overall_percentiles[fifa_overall_percentiles['Known As'] == best_player]
attribute_values2 = player_row2.iloc[:, 1:].values.flatten()

# Print the attribute values

attribute_values2    #percentile overall
```

```python
ratio = np.divide(attribute_values2, attribute_values1)
#ratio
```

```python
# Create a new array with modified values
ratio_new = np.where(ratio < 0.99, 0, ratio)

# Display the new array
#print(ratio_new)
```

```python
# Reshape attribute_values to match the shape of the DataFrame
attribute_values_reshaped = attribute_values1.reshape(1, -1)

# Create a DataFrame for Messi's attribute percentiles
messi_df = pd.DataFrame(attribute_values_reshaped, columns=filtered_df.columns[1:])
```

```python
#messi_df
```

```python
# Create a new DataFrame to store the differences
difference_df = pd.DataFrame()

# Iterate over the rows of the filtered DataFrame
for index, row in filtered_df.iterrows():
    player_name = row['Known As']
    player_percentiles = row.iloc[1:]
    
    # Calculate the difference between player percentiles and Messi's percentiles
    difference =  messi_df.values[0] - player_percentiles
    
    # Add the difference values to the difference DataFrame along with the player name
    difference_df = difference_df.append(pd.Series([player_name] + difference.tolist(), index=['playername'] + messi_df.columns.tolist()), ignore_index=True)
```

```python
#difference_df
```

```python
df_result = difference_df.copy()
df_result.iloc[:, 1:] = df_result.iloc[:, 1:].multiply(ratio_new, axis=1)
#df_result
```

```python
# Create a new Excel application
app = xw.App(visible=True)

# Create a new workbook
wb = app.books.add()

# Select the first sheet
sheet = wb.sheets[0]

# Write the DataFrame to the sheet, including column names
sheet.range("A1").value = df_result.columns.tolist()
sheet.range("A2").value = df_result.values.tolist()

# Calculate the sum of values for each row
sum_formula = "=SUM(B{row}:BN{row})"
sum_column = sheet.range((1, df_result.shape[1] + 1))
sum_column.value = "Sum"
for row in range(2, df_result.shape[0] + 2):
    sum_formula_row = sum_formula.format(row=row)
    sheet.range((row, sum_column.column)).formula = sum_formula_row

# Calculate the last row in the sum column
last_row = df_result.shape[0] + 1

# Apply bold font to the header cells
header_range = sheet.range("A1").expand("right")
header_range.api.Font.Bold = True

# Set the default column width to 15
sheet.api.Columns.ColumnWidth = 15

# Add a filter to the sum column
sum_column.api.AutoFilter(1)

# Sort column BO in ascending order
sort_column = sheet.range("BO2:BO{}".format(last_row))
sort_column.sort()

# Save the workbook to an Excel file
wb.save("output.xlsx")








```
